```python
import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt
```

```python
# 1.1 Missing Values
# 1.1.1 Problem: How to handle missing values in a DataFrame?
```

```python
df = pd. DataFrame({'A': [1, 2, 3, None, 5],
                    'B': [None, 6, None, 8, 9],
                    'C': [10, 11, 12, None, 14]})
```

```python
df.fillna(df.mean(numeric_only=True), inplace=True)
```

```python
df
```

```python
df.plot()
plt.show()
```

```python
# 1.2 Maintaining DataFrame Data
# 1.2.1 Problem: How to drop duplicate rows in a DataFrame?
```

```python
df = pd.DataFrame({'A': [1, 1, 2, 3, 3],
                   'B': [4, 5, 6, 7, 7],
                   'C': [8, 9, 9, 10, 10]})
```

```python
df.drop_duplicates(inplace=True)
```

```python
df
```

```python
df.plot()
plt.show()
```

```python
# 1.3 Joining Two DataFrames
# 1.3.1 Problem: How to join two DataFrames based on a common column?
```

```python
df1 = pd.DataFrame({'key': ['A', 'B', 'C', 'D'],
                    'value': [1, 2, 3, 4]})
```

```python
df2 = pd.DataFrame({'key': ['C', 'D', 'E', 'F'],
                    'value': [5, 6, 7, 8]})
```

```python
df1
```

```python
df2
```

```python
df_merged = pd.merge(df1, df2, on='key')
```

```python
df_merged
```

```python
df_merged.plot()
plt.show()
```

```python
# 2.1 Working with Dates
# 2.1.1 Problem: How to extract year, month, and day from a
# date column in a DataFrame?
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/finance-charts-apple.csv')
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df['Year'] = df['Date'].dt.year
```

```python
df['Month'] = df['Date'].dt.month
```

```python
df['Day'] = df['Date'].dt.day
```

```python
df.plot(x='Date', y='AAPL.Adjusted')
plt.show()
```

```python
# 2.2 Data Aggregation
# 2.2.1 Problem: How to group a DataFrame by a column and
# calculate summary statistics?
```

```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips_group = tips.groupby('day')['total_bill'].agg(['mean', 'std'])
```

```python
tips_group.plot(kind='bar')
plt.ylabel('Tips')
plt.show()
```

```python
# 2.3 Data Visualization
# 2.3.1 Problem: How to create a histogram from a column in a DataFrame?
```

```python
tips['total_bill'].hist(bins=20)
plt.show()
```

```python
# 2.4 Data Transformation
# 2.4.1 Problem: How to create a new column in a DataFrame based on conditions?
```

```python
tips['tip_category'] = 'low'
```

```python
tips.loc[tips['tip'] >= 5, 'tip_category'] = 'high'
```

```python
tips.plot(kind='scatter',
          x='total_bill',
          y='tip',
          c=tips['tip_category'].map({'low': 'orange', 'high': 'green'}))
plt.show()
```
