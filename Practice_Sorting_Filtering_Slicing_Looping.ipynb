```python
# 1.9 Load the "iris" dataset into a pandas DataFrame and create a
# scatter plot showing the relationship between the "sepal_length"
# and "sepal_width" columns, with different colors for each species of iris.
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
df_iris = pd.DataFrame(data=iris['data'], columns=iris['feature_names'])
```

```python
df_iris['target'] = iris['target']
```

```python
colors = ['red', 'green', 'blue']
```

```python
for i in range(len(iris['target_names'])):
    plt.scatter(df_iris[df_iris['target'] == i]['sepal length (cm)'],
                df_iris[df_iris['target'] == i]['sepal width (cm)'],
                c=colors[i],
                label=iris['target_names'][i])
plt.legend()
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.show()
```
